### Create Character Vector Database

```python
from langchain.text_splitter import RecursiveCharacterTextSplitter
from langchain_cohere import CohereEmbeddings
import json
import random
from langchain.vectorstores import Chroma

# Set the game name and character name
game_name = 'God_of_War_2018'
character_name = 'Atreus'

# Load the file
with open(f'{game_name}/characters/{character_name}/character_knowledge.txt', errors='ignore') as f:
    public_info = f.read()

# Create Text Splitter
text_splitter = RecursiveCharacterTextSplitter(
    chunk_size=600,
    chunk_overlap=0,
    length_function=len,
    separators=['\n\n\n', '\n\n', '\n', '.', ',', ' ']
)

# Split the text
public_info_list = text_splitter.split_text(public_info)

# Define your API key and model
selected_key = '6LQRTglHmlVTGr1Se2ezKZvP4rU89jYDclfhi5tR'
model_name = 'large'  # Replace with the appropriate model name from the Cohere documentation

print(f"Using API Key: {selected_key}")

# Initialize the embedding model
try:
    embeddings = CohereEmbeddings(cohere_api_key=selected_key, model=model_name)
    print("Embeddings model initialized successfully.")
except KeyError as e:
    print(f"KeyError: {e}. Ensure that the API key is valid and check library documentation.")
    embeddings = None
except Exception as e:
    print(f"An error occurred: {e}.")
    embeddings = None

# Only proceed if embeddings were initialized successfully
if embeddings is not None:
    # Path for the vector database
    persist_directory = f'{game_name}/characters/{character_name}/vectordb'
    
    # Create the Vector Database
    vectordb = Chroma.from_texts(public_info_list, embeddings, persist_directory=persist_directory)
    vectordb.persist()
else:
    print("Failed to initialize embeddings. Vector database creation aborted.")
```

### Test the Vector Database

```python
# search for a topic and see if you get relevant information, k is for number of similar items to return
docs = vectordb.similarity_search("Loki", k = 3)
similar_story_string = "\n".join(doc.page_content for doc in docs)
print(similar_story_string)
```
