### Place 5 images in the Game/characters/character_name/images folder from 1.jpg to 5.jpg and one image in root directory named as test.jpg

```python
from deepface import DeepFace

# Set the game name and character name
game_name = 'God_of_War_2018'
character_name = 'Atreus'

dfs = DeepFace.find(img_path = "test.jpg", db_path = f"{game_name}/characters/{character_name}/images", model_name='Facenet')
```

```python
if not dfs.empty:
    # Access the first row and the columns 'identity' and 'Facenet_cosine'
    best_image_path = dfs.iloc[0]['identity']
    best_facenet_cosine_score = dfs.iloc[0]['Facenet_cosine']
    
    print("Best Image path:", best_image_path)
    print("Best Facenet_cosine score:", best_facenet_cosine_score)
else:
    print("No matching images found in the database.")
```
